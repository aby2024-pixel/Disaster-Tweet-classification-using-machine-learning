```python
# Disaster Tweet Classification - Full Script for Provided CSV Files

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score, roc_auc_score
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
torch.manual_seed(42)

# Load Data 
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

print(f"Train shape: {df_train.shape}, Test shape: {df_test.shape}")

# Preprocessing 
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    text = ' '.join(text.split())
    return text

df_train['cleaned_text'] = df_train['text'].apply(clean_text)
df_test['cleaned_text'] = df_test['text'].apply(clean_text)

# Exploratory Data Analysis 
print("Class balance:\n", df_train['target'].value_counts())
plt.figure(figsize=(8,5))
sns.countplot(data=df_train, x='target')
plt.title('Class Distribution')
plt.xlabel('Target (0=Non-Disaster, 1=Disaster)')
plt.savefig('eda_class_distribution.png', dpi=150)
plt.close()

df_train['text_length'] = df_train['text'].apply(len)
plt.figure(figsize=(10,5))
df_train[df_train['target']==0]['text_length'].hist(bins=40, alpha=0.6, label='Non-Disaster', color='blue')
df_train[df_train['target']==1]['text_length'].hist(bins=40, alpha=0.6, label='Disaster', color='red')
plt.xlabel('Tweet Length')
plt.ylabel('Frequency')
plt.title('Tweet Length by Class')
plt.legend()
plt.savefig('eda_text_length.png', dpi=150)
plt.close()

# Train/Validation Split 
X_train, X_val, y_train, y_val = train_test_split(
    df_train['cleaned_text'], df_train['target'], test_size=0.2, random_state=42, stratify=df_train['target']
)

# Baseline Model
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_val_tfidf = tfidf_vectorizer.transform(X_val)

baseline_model = LogisticRegression(max_iter=1000, random_state=42)
baseline_model.fit(X_train_tfidf, y_train)

y_pred_baseline = baseline_model.predict(X_val_tfidf)
y_pred_proba_baseline = baseline_model.predict_proba(X_val_tfidf)[:, 1]

print("\n=== Baseline Model ===\n", classification_report(y_val, y_pred_baseline))
print("Accuracy:", accuracy_score(y_val, y_pred_baseline))
print("F1-Score:", f1_score(y_val, y_pred_baseline))
print("ROC-AUC:", roc_auc_score(y_val, y_pred_proba_baseline))

cm_baseline = confusion_matrix(y_val, y_pred_baseline)
plt.figure(figsize=(8,6))
sns.heatmap(cm_baseline, annot=True, fmt='d', cmap='Blues')
plt.title('Baseline Confusion Matrix')
plt.savefig('baseline_confusion_matrix.png', dpi=150)
plt.close()

# DistilBERT Model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
def tokenize_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=128)

train_dataset = Dataset.from_dict({'text': X_train.tolist(), 'label': y_train.tolist()})
val_dataset = Dataset.from_dict({'text': X_val.tolist(), 'label': y_val.tolist()})

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)
train_dataset.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
val_dataset.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])

model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)
training_args = TrainingArguments(
    output_dir='./results', num_train_epochs=3, per_device_train_batch_size=16, 
    per_device_eval_batch_size=32, warmup_steps=500, weight_decay=0.01,
     learning_rate=2e-5, seed=42
)
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {'accuracy': accuracy_score(labels, preds), 'f1': f1_score(labels, preds)}
trainer = Trainer(model=model, args=training_args, train_dataset=train_dataset, eval_dataset=val_dataset, compute_metrics=compute_metrics)
trainer.train()

eval_results = trainer.evaluate()
print("\n=== DistilBERT Results ===")
print("Validation Loss:", eval_results['eval_loss'])
print("Accuracy:", eval_results['eval_accuracy'])
print("F1-Score:", eval_results['eval_f1'])

distilbert_predictions = trainer.predict(val_dataset)
y_pred_dl = np.argmax(distilbert_predictions.predictions, axis=-1)
cm_dl = confusion_matrix(y_val, y_pred_dl)
plt.figure(figsize=(8,6))
sns.heatmap(cm_dl, annot=True, fmt='d', cmap='Greens')
plt.title('DistilBERT Confusion Matrix')
plt.savefig('distilbert_confusion_matrix.png', dpi=150)
plt.close()

# prediction function
def predict_tweet(text, model, tokenizer):
    cleaned = clean_text(text)
    inputs = tokenizer(cleaned, return_tensors="pt", truncation=True, padding='max_length', max_length=128)
    # Move tensors to model device!
    device = next(model.parameters()).device
    for k in inputs:
        inputs[k] = inputs[k].to(device)

    model.eval()
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        probabilities = torch.softmax(logits, dim=1)
        predicted_class = torch.argmax(probabilities, dim=-1).item()
        confidence = probabilities[0][predicted_class].item()
    label_mapping = {0: 'Non-Disaster', 1: 'Disaster'}
    predicted_class = label_mapping[predicted_class]
    print(f"Tweet: {text}\n")
    print(f"Predicted Class: {predicted_class}\n")
    print(f"Confidence: {confidence:.2f}\n")
    return predicted_class


predict_tweet("Earthquake just hit! Building shaking, people running outside #emergency", model, tokenizer)

print("\n finished. All the outputs are saved and printed as described.")
```
